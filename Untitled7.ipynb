```python
import matplotlib.pyplot as plt
import numpy as np
import math as mt
from mpl_toolkits.mplot3d import Axes3D
```

```python
Q = 5e14
E = 180e3
T = 1373
p = 2
k = 8.62e-5
N0 = 1e16
```

```python
def Dif(C, temperature, ni):
    Nc = 6.2e15 * temperature ** 1.5
    Nv = 3.5e15 * temperature ** 1.5
    Eg = 1.17 - 4.73e-4 * (temperature ** 2) / (temperature + 636)
    ni = ((Nc * Nv / 2) ** 0.5) * mt.exp(-Eg / (2 * k * temperature))
    
    return 0.214 * mt.exp(-3.65/(k*temperature)) + 15 * C / ni * mt.exp(-4.08/(k*temperature))
```

```python
def implantation(x, y, Q, dRp, Rp, dRpl, a):
    return Q / np.sqrt(2 * np.pi) / dRp / 1e-4 * np.exp(-(x - Rp) ** 2 / 2 / dRp ** 2) / 2 * (
                mt.erfc((y - a) / np.sqrt(2) / dRpl) - mt.erfc((y + a) / np.sqrt(2) / dRpl))
```

```python
class Si:
    def __init__(self):
        # set constants
        self.m_dn = 1.08
        self.m_dp = 0.59
        self.epsilon = 11.7
        self.mu_n_300 = 1350
        self.mu_p_300 = 450
        
    def Eg(self, temperature):
        return 1.17 - 4.73e-4 / (temperature + 636) * temperature ** 2
 
 
    def Ei(self, temperature):
        return self.Eg(temperature) / 2 - k * temperature / 4
 
    def Nv(self, temperature):
        return 4.82e15 * (self.m_dp ** 1.5) * (temperature ** 1.5)
 
    def Nc(self, temperature):
        return 4.82e15 * (self.m_dn ** 1.5) * (temperature ** 1.5)
 
    def ni(self, temperature):
        return ((self.Nc(temperature) * self.Nv(temperature)) ** 0.5) * mt.exp(
            -1 * self.Eg(temperature) / (2 * temperature * k))
 
    def Rp(self, E):
        return self.a1 * E ** self.a2 + self.a3
 
    def dRp(self, E):
        return self.a4 * E ** self.a5 + self.a6
 
    def gamma(self, E):
        return self.a7 / (self.a8 + E) + self.a9
 
    def betta(self, E):
        return self.a10 / (self.a11 + E) + self.a12 + self.a13 * E
 
    def dRpl(self, E):
        return np.sqrt(2 * self.Rp(E) / (self.Se + self.Sn) * E - self.dRp(E) ** 2 - self.dRp(E) ** 2)
 
    def C_minus(self, temperature):
        return mt.exp((self.Ei(temperature) + 0.57 - self.Eg(temperature)) / k / temperature)
 
    def C_plus(self, temperature):
        return mt.exp((0.35 - self.Ei(temperature)) / k / temperature)
 
    def C_doubleminus(self, temperature):
        return mt.exp((2 * self.Ei(temperature) + 1.25 - 3 * self.Eg(temperature)) / k / temperature)
 
    def gamma_l(self, temperature):
        return 2620 * mt.exp(-1.1 / k / temperature)
 
    def gamma_p(self, temperature):
        return 9.63e-16 * mt.exp(2.83 / k / temperature)
 
    def Vn(self, temperature, n):
        return (1 + self.C_plus(temperature) * self.ni(temperature) / n
                + self.C_minus(temperature) * n / self.ni(temperature)
                + self.C_doubleminus(temperature) * (n / self.ni(temperature)) ** 2) \
               / (1 + self.C_minus(temperature) + self.C_plus(temperature) + self.C_doubleminus(temperature))
```

```python
def C_time(y, n, dx, T, ni):
    a = np.empty(n)
    b = np.empty(n)
    d = np.empty(n)
    r = np.empty(n)
    delta = np.empty(n)
    lyamda = np.empty(n)
    if n == 40:
        d[0] = 1
        a[0] = -1
    else:
        d[0] = 0
        a[0] = 1
    b[0] = 0
    r[0] = 0
    d[-1] = 0
    a[-1] = 1
    b[-1] = 0
    r[-1] = 0
    delta[0] = -d[0]/a[0]
    lyamda[0] = r[0]/a[0]
    dt = 1
    for i in range(1, n-1, 1):
        a[i] = -(2 + (dx ** 2 * 1e-8) / (Dif(y[i], T, ni) * dt))
        r[i] = -(((dx ** 2 * 1e-8) * y[i]) / (Dif(y[i], T, ni) * dt))
        b[i] = 1
        d[i] = 1
    for i in range(1, n, 1):
        delta[i] = -d[i] / (a[i] + b[i] * delta[i - 1])
        lyamda[i] = (r[i] - b[i] * lyamda[i - 1]) / (a[i] + b[i] * delta[i - 1])
    y[-1] = lyamda[-1]
    for i in range(n - 2, -1, -1):
        y[i] = delta[i] * y[i + 1] + lyamda[i]
    return y
```

```python
class Stibium(Si):
    def __init__(self):
       # stibium
        self.a1 = 0.000688
        self.a2 = 0.983
        self.a3 = 0.003962
        self.a4 = 0.000402
        self.a5 = 0.874
        self.a6 = 0.000582
        self.a7 = 339.8
        self.a8 = 342
        self.a9 = -0.5051
        self.a10 = 38.73
        self.a11 = 61.7
        self.a12 = 2.559
        self.a13 = 0
 
        self.Se = 330
        self.Sn = 15
        self.m_dn = 1.08
        self.m_dp = 0.59
        self.epsilon = 11.7
        self.mu_n_300 = 1350
        self.mu_p_300 = 450
Stibium = Stibium()
 
Si = Si()
```

```python
def oxidation(time, T, concentration, xi):
    sc = Si
    B = K(12.9, 1.23, T) * (1 + sc.gamma_p(T) * sc.ni(T) ** 0.22)
    A = B / (K(1.04e5, 2, T) * (1 + sc.gamma_l(T) * (sc.Vn(T, concentration) - 1)))
    # вызов функции расчета толщины оксисла с новыми параметрами
    x = x_t(time, A, B, xi)
    return x
```

```python
def K(A, Ea, temperature):
    return A * mt.exp(- Ea / k / temperature)
```

```python
def x_t(t, A, B, xi):
    tau = (xi ** 2 + A * xi) / B
    return (A / 2) * (mt.sqrt(1 + (t + tau) / A ** 2 * 4 * B) - 1)
```

```python
def oxide(T, time, concentration, x0):
    sc = Si
    n = time * 60
    t = np.empty(n)
    x = np.empty(n)
    t[0] = 0
    x[0] = x0
    if time == 1:
        x0 = 50e-3
    else:
        x0 = x0
    xi = x0 * 2.43
    for i in range(1, n):
        t[i] = t[i - 1] + 1
        x[i] = oxidation(t[i], T, concentration, x0)
    return t, x, x[-1]
```

```python
class Plot:
    Y = []
    X = []
    ox_plot_x = []
    ox_plot_y = []
    z_ox = []
    def draw(self, start, end, x, y, dx, dy, xmax, T, Rp, dRp, dRpl, Q, ni, t):
        for i in range(len(x)-1, 0, -1):
            if i % 2 == 0:
                self.Y.append(-2*x[i])
        for j in range(0, len(x), 1):
            self.X.append(x[j])
            if j % 2 == 0:
                self.Y.append(2*x[j])
        dy = dx
        X, Y = np.meshgrid(self.X, self.Y)
        Z = X ** 2 - Y ** 2
        for i in range(0, len(self.Y), 1):
            for j in range(0, len(self.X), 1):
                Z[i][j] = implantation(X[i][j], Y[i][j], Q, dRp, Rp, dRpl, xmax / 2)
        fig = plt.figure('graph')
        ax = fig.add_subplot(projection='3d')
        fig.suptitle('Концентрация сурьмы')
 
        def get_junct(x, y ,t, end, conc = 1e16):
            X_r = X.copy()
            Y_r = Y.copy()
            Z_r = Z.copy()
            x_new = x
            y_new = y
            z_new = np.zeros(end)
            junc_arr = np.zeros(end)
            index = []
            for j in range(1, t, 1):
                for i in range(0, len(Z_r[1, :]), 1):
                    Z_r[:, i] = C_time(Z_r[:, i], end - 1, dx, T, ni)
                for i in range(0, len(Z[:, 1]), 1):
                    Z_r[i] = C_time(Z_r[i], end, dy, T, ni)
            for i in range(0, end-1):
                junc_arr[i] = Z_r[int(end / 2)][i]
                if (junc_arr[i] > conc and junc_arr[i - 1] < conc) or (junc_arr[i] < conc and junc_arr[i - 1] > conc):
                    index.append(i - 1)
            for i in range(0, end - 1):
                z_new[i]= (Z_r[i])
 
            fig = plt.figure('Сечение p-n перехода')
            fig.suptitle('Сечение p-n перехода')
            plt.xlabel('y, мкм')
            plt.ylabel('Концентрация, см-3')
            plt.plot(y, z_new)
            plt.show()
 
        def spread(t, y, end, data1):
            tox = np.zeros(end)
            for j in range(0, end):
                if y[j] > 1 and y[j] < 3:
                    tox[j] = data1[-1]
                else:
                    if y[j] < 1:
                        tox[j] = data1[-1] / 2 * (1 + 1 \
                             / (1 + 0.27893 * 2 ** 0.5 / 2 \
                                 * (1 - y[j]) / data1[-1] \
                                 + 0.230389 * (2 ** 0.5 / 2 * (1 - y[j]) \
                                               / data1[-1]) ** 2 + 0.000972 *
                                 (2 ** 0.5 / 2 * (1 - y[j]) / data1[-1]) ** 3 \
                                 + 0.078108 * (2 ** 0.5 / 2 * (1 - y[j]) / data1[-1]) ** 4) ** 4)
                    else:
                        tox[j] = data1[-1] / 2 * (1 + 1 \
                                / (1 + 0.27893 * 2 ** 0.5 / 2 \
                                   * (y[j] - 3) / data1[-1] \
                                   + 0.230389 * (2 ** 0.5 / 2 * (y[j] - 3) \
                                                 / data1[-1]) ** 2 + 0.000972 *
                                   (2 ** 0.5 / 2 * (y[j] - 3) / data1[-1]) ** 3 \
                                   + 0.078108 * (2 ** 0.5 / 2 * (y[j] - 3) / data1[-1]) ** 4) ** 4)
            return y, tox
 
        def get_Z(t, end):
            X_r = X.copy()
            Y_r = Y.copy()
            Z_r = Z.copy()
            for j in range(1, t, 1):
                for i in range(0, len(Z_r[1, :]), 1):
                    Z_r[:, i] = C_time(Z_r[:, i], end - 1, dx, T, ni)
                for i in range(0, len(Z[:, 1]), 1):
                    Z_r[i] = C_time(Z_r[i], end, dy, T, ni)
            index = int(end / 2)
            return Z_r[index][index]
 
        def update(t, end):
            ax.clear()
            X_r = X.copy()
            Y_r = Y.copy()
            Z_r = Z.copy()
            ax.set_xlim(0, xmax)
            ax.set_ylim(-2 * xmax, 2 * xmax)
            ax.set_zlim(0, 3e19)
            ax.set_xlabel('x, мкм')
            ax.set_ylabel('y, мкм ')
            ax.set_zlabel('Концентрация')
            t = int(t)
            for j in range(1, t, 1):
                for i in range(0, len(Z_r[1, :]), 1):
                    Z_r[:, i] = C_time(Z_r[:, i], end - 1, dx, T, ni)
                for i in range(0, len(Z[:, 1]), 1):
                    Z_r[i] = C_time(Z_r[i], end, dy, T, ni)
 
            surface2 = ax.plot_wireframe(X_r, Y_r, Z_r, rstride=1, cstride=1, color='yellow', linewidth=0.5, label='После')
 
            surface = ax.plot_wireframe(X, Y, Z, rstride=1, cstride=1, color='black', linewidth=0.5, label='До разгонки')
            ax.legend(title="Концентрации")
        update(t, end)
        x0_arr = [50e-3]
        for i in range(1, t):
            if i == 1:
                x0 = x0_arr[0]
            else:
                x0 = x0_arr[i - 1]
            data = oxide(T, i, get_Z(i, end), x0)
            x0_arr.append(data[2])
        get_junct(x, y, t, end)
        max_ox = max(data[1])
        spread_data = spread(t * 60, y, end, data[1] )
        fig = plt.figure('spread')
        plt.plot(spread_data[0], spread_data[1])
        plt.xlabel("у, мкм")
        plt.ylabel('Толщина окисла, мкм')
        plt.show()
    plt.show()
 
Plot = Plot()
```

```python
def fulldesp(Q, E, T, t):
    sc = Stibium
    Rp = sc.Rp(E)
    dRp = sc.dRp(E)
    dRpl = sc.dRpl(E)
    ni = sc.ni(T)
    n = 70
    xmax = 1
    ymax = 4
    dx = xmax / n
    dy = ymax / n
    y = np.empty(n)
    x = np.empty(n)
    for i in range(0, n):
        if i == 0:
            x[i] = 0
            y[i] = 0
        else:
            x[i] = x[i - 1] + dx
            y[i] = y[i - 1] + dy
    Plot.draw(0, n, x, y, dx, dy, xmax, T, Rp, dRp, dRpl, Q, ni, t)   
```
